# Importación de librerias

```python
import numpy as np
import vedo
import trimesh
import json
```

# Datos estructurados (base de datos)

```python
objects_data = [
    {"type": "sphere",   "pos": [0,  0, 0],  "scale": 0.5, "color": "red"},
    {"type": "cube",     "pos": [2,  0, 0],  "scale": 0.8, "color": "blue"},
    {"type": "cylinder", "pos": [4,  0, 0],  "scale": 0.6, "color": "green"},
    {"type": "sphere",   "pos": [0,  2, 0],  "scale": 1.0, "color": "orange"},
    {"type": "cube",     "pos": [2,  2, 0],  "scale": 0.4, "color": "purple"},
    {"type": "cylinder", "pos": [4,  2, 0],  "scale": 0.9, "color": "yellow"},
    {"type": "sphere",   "pos": [0,  4, 0],  "scale": 0.7, "color": "cyan"},
    {"type": "cube",     "pos": [2,  4, 0],  "scale": 1.2, "color": "pink"},
    {"type": "cylinder", "pos": [4,  4, 0],  "scale": 0.3, "color": "tomato"},
]
```

### Se procede a generar con `vedo` y exportarlos en una vista

```python
def build_scene(data):
    meshes = []
    for obj in data:
        x, y, z = obj["pos"]
        scale = obj["scale"]
        color = obj["color"]

        if obj["type"] == "sphere":
            m = vedo.Sphere(pos=(x, y, z), r=scale).color(color)

        elif obj["type"] == "cube":
            m = vedo.Box(pos=(x, y, z), width=scale, height=scale).color(color)

        elif obj["type"] == "cylinder":
            m = vedo.Cylinder(pos=(x, y, z), r=scale*0.5, height=scale).color(color)

        meshes.append(m)

    return meshes

scene_meshes = build_scene(objects_data)

plotter = vedo.Plotter(title="Escena Paramétrica")
plotter.show(*scene_meshes, axes=1)
```

### Ahora se exporta cada malla a `.obj` usando `trimesh`

```python
import os
os.makedirs("exports", exist_ok=True)

for i, (obj, vmesh) in enumerate(zip(objects_data, scene_meshes)):
    tri_mesh = vmesh.triangulate()
    
    verts = np.array(tri_mesh.points)
    cells = tri_mesh.cells 
    
    faces = np.array([c for c in cells if len(c) == 3])
    
    tm = trimesh.Trimesh(vertices=verts, faces=faces, process=False)
    
    filename = f"exports/obj_{i:02d}_{obj['type']}.obj"
    trimesh.exchange.export.export_mesh(tm, filename)
    print(f"Exportado: {filename}")
```

# BONUS: Leer desde JSON y regenerarlo

```python
# Guardar como JSON
with open("exports/scene_data.json", "w") as f:
    json.dump(objects_data, f, indent=2)

# Regenerar desde JSON
with open("exports/scene_data.json") as f:
    loaded_data = json.load(f)

scene_from_json = build_scene(loaded_data)
plotter2 = vedo.Plotter(title="Escena desde JSON")
plotter2.show(*scene_from_json, axes=1)
```
