```python
from diffusers import StableDiffusionPipeline
import torch
import os

pipe = StableDiffusionPipeline.from_pretrained(
 "runwayml/stable-diffusion-v1-5",
 torch_dtype=torch.float16
).to("cuda")
```

```python
prompt = "A surreal futuristic city in the clouds, digital art"

generator = torch.Generator("cuda").manual_seed(42)
image = pipe(
    prompt,
    num_inference_steps=50,
    guidance_scale=7.5,
    generator=generator
).images[0]
image.save("media/inference_steps.png")
print("Guardada: media/inference_steps.png")
```

```python
positive_prompt = "A beautiful mountain landscape with a river, sunny, photorealistic, 4k"
negative_prompt = "blurry, low quality, ugly, deformed, watermark, text, people, buildings"

generator = torch.Generator("cuda").manual_seed(42)
image = pipe(
    positive_prompt,
    negative_prompt=negative_prompt,
    num_inference_steps=35,
    guidance_scale=7.5,
    generator=generator
).images[0]
image.save("media/negative_prompt.png")
print("Guardada: media/negative_prompt.png")
```
